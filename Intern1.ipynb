```python
#Importing all necessary packages
import pandas as pd
```

```python
#Reading the training data
df=pd.read_csv("train.csv")
df.head()
```

```python
#Showing all the features present in the dataset
df.columns
```

```python
#Setting all relevent data to the Linear Regression including output feature inside another variable
d=df[['SalePrice','LotArea','BedroomAbvGr','BsmtFinSF1','BsmtFinSF2','MasVnrArea','BsmtUnfSF','TotalBsmtSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','GarageArea','PoolArea','WoodDeckSF','OpenPorchSF']]
d.head()
```

```python
#Printing shape of the newly made variable
d.shape
```

```python
#Checking that if there is any null value present in the dataset for any feature
d.isnull().sum()
```

```python
#Discarding the rows consisting of null values in any of the feature 
d.dropna(inplace=True)
d.shape
```

```python
#Defining the values of x_train and y_train  from the training dataset
x_train=d.drop('SalePrice',axis=1)
y_train=d['SalePrice']
```

```python
#Fitting the training data into the linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Printing the first 5 values ofSalePrice columns using "head" function
d["SalePrice"].head()
```

```python
#Predicting the values of output variable using the built model
y_predict=lr.predict(d.iloc[:10,1:])
y_predict
```

```python
#Finding the model accuracy from training data
lr.score(x_train,y_train)
```
